```python
%%execsql

CREATE TABLE nate.ewhg_materials AS 

SELECT DISTINCT material
from (
    SELECT DISTINCT MATERIAL 
    FROM nate.simple_features
    WHERE plant in( 'EWHG','HCS0') 
UNION ALL
      SELECT DISTINCT COMPonent as material
      from nate.simple_features
    WHERE plant in( 'EWHG','HCS0') 
) a
DISTRIBUTED BY(material);
    
```
