# Now we'll use the service orders, list of components for each service order, equipment master and material master to create a single table with much of the information that will be valueable to us:

```python
%%execsql
DROP TABLE IF EXISTS nate.simple_features;
CREATE TABLE nate.simple_features AS 
    (
    SELECT a.ordr, 
        case when asong.is_numeric(c.material) then trim(leading '0' from c.material) else c.material end as material
        , e.txtmd as material_desc, a.material AS component, d.txtmd as component_desc, a.qtywithdrawn, a.uom, a.plant
    FROM data_science.ordercomponents_int a 
    JOIN sap_bi.bic_azcs_o0100_int b
    ON a.ordr = trim(leading '0' from b.cs_order)
    JOIN sap_bi.bic_azequip00_int c
    ON c.equipment = b.equipment
    JOIN sap_bi.bi0_tmaterial_int d
    ON d.material = a.material
    JOIN sap_bi.bi0_tmaterial_int e
    ON e.material = c.material
    WHERE a.qtywithdrawn > 0 and a.material is not null and char_length(a.material) > 0 and d.langu = 'E' and e.langu = 'E'
    )
DISTRIBUTED BY (ordr);
```

Now we have matched up every component withdrawn for a service order to the material it was withdrawn for, along with descriptions for both the material and the component.

```python
%%showsql 
SELECT material, material_desc, component, component_desc, sum(qtywithdrawn) as qty, count(distinct ordr) as orders, 
sum(qtywithdrawn)/count(distinct ordr) as qtyperorder
FROM nate.simple_features 
WHERE uom IN ('EA','PCE')
GROUP BY 1,3,2,4
ORDER BY 7 DESC
LIMIT 10;
```

Here's a table listing the total number of components issued for each material, along with the total number of unique components for each:

```python
%%showsql
SELECT material, count(distinct component) 
FROM nate.simple_features 
GROUP BY 1
ORDER BY 2 DESC
LIMIT 100
```

```python
%%showsql
SELECT material, count(distinct component) as uniquecount, count(component) as allcount
FROM nate.simple_features 
GROUP BY 1
ORDER BY 2 DESC,3 DESC
LIMIT 100
```

Table with material matched to an array of every past component that has been used to service that material:

```python
%%showsql
SELECT material, array_agg(distinct component) as component_list
FROM nate.simple_features 
GROUP BY 1
LIMIT 100
```

```python
%%execsql
DROP TABLE IF EXISTS nate.material_ordrcount;
CREATE TABLE nate.material_ordrcount AS (SELECT material, count(distinct ordr) as ordrtot
FROM nate.simple_features
GROUP BY 1)
```

# First weak predictive calculations - empirically  calculated probability
# For each material-component pair, this table lists the probability based on past data that the listed component will be needed to service the listed material

```python
%%showsql
SELECT a.material, component, count(distinct ordr)/b.ordrtot::float as compprob
FROM nate.simple_features a, nate.material_ordrcount b
WHERE a.material = b.material
GROUP BY 1,2,b.ordrtot
LIMIT 100
```
