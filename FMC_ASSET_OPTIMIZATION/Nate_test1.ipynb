```python
sql = """ 
SELECT twenty_news_groups_term_frequencies.token, twenty_news_groups_term_frequencies.doc_id 
FROM nate.twenty_news_groups_vocabulary, nate.twenty_news_groups_term_frequencies 
WHERE twenty_news_groups_vocabulary.token = twenty_news_groups_term_frequencies.token
ORDER BY twenty_news_groups_term_frequencies.doc_id
"""
df = psql.read_sql(sql, conn)
conn.commit()
display(df)
```

```python
sql = """
SELECT doc_id, count(twenty_news_groups_term_frequencies.token)
FROM nate.twenty_news_groups_vocabulary, nate.twenty_news_groups_term_frequencies 
WHERE twenty_news_groups_vocabulary.token = twenty_news_groups_term_frequencies.token
GROUP BY doc_id
ORDER BY doc_id
"""
df = psql.read_sql(sql, conn)
conn.commit()
display(df)
```
